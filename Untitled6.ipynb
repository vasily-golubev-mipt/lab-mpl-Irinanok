```python
import matplotlib.pyplot as plt
import math

txt = str(input())
text = open(txt, 'r')
N = int(text.readline())
x = []
y = []
for i in range(N):
    m = []
    m.extend(map(float, text.readline().split()))
    x.append(m[0])
    y.append(m[1])
c = 'Number of points: ' + str(len(x))
min_x = math.floor(min(x)) - 1
min_y = math.floor(min(y)) - 1
max_x = math.ceil(max(x)) + 1
max_y = math.ceil(max(y)) + 1
size = int(input())
plt.scatter(x, y, c = 'b', s = size)
plt.xlim(min_x, max_x)
plt.ylim(min_y, max_y)
plt.axis('scaled')
plt.suptitle(c)
plt.show()
```

```python
import matplotlib.pyplot as plt
txt = str(input())
file = open(txt, 'r')
frames = []
N = int(sum(1 for i in file))
file.seek(0)
min_x = 0
min_y = 0
max_x = 0
max_y = 0

for i in range(int(N/2)):
    x = [float(x) for x in file.readline().split()]
    y = [float(y) for y in file.readline().split()]
    min_x = min(min_x, min(x))
    min_y = min(min_y, min(y))
    max_x = max(max_x, max(x))
    max_y = max(max_y, max(y))
    frames.append([x, y])
max_range_y = max(max_y, min_y * (-1))
fig, axs = plt.subplots(int(N / 6), 3, figsize=(15, 10))

for i in range(0, int(N / 2)):
    axs[int(i/3), i%3].plot(frames[i][0], frames[i][1])
    axs[int(i/3), i%3].title.set_text("Frame " + str(i + 1))
    axs[int(i/3), i%3].grid()

plt.setp(axs, xlim=(min_x, max_x), ylim=(max_range_y * (-1), max_range_y))
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
import numpy as np


def sum_arr(k):
    sum_prep = 0.0
    sum_group = 0.0
    for a in range(k):
        sum_prep += prep_mark[:, a]
        sum_group += group_mark[:, a]
    return (sum_prep, sum_group)


marks = []
group = []
prep = []

colors = ["#191970", "#0000CD", "#4169E1", "#1E90FF", "#00BFFF", "#00FFFF", "#7FFFD4", "#20B2AA"]
txt = str(input())
with open(txt, 'r') as file:
    csv_read = csv.reader(file, delimiter=';')
    for student in csv_read:
        prep.append(str(student[0]))
        group.append(str(student[1]))
        marks.append(int(student[2]))

prep_mark = np.array([np.zeros(8) for i in range(7)])
group_mark = np.array([np.zeros(8) for i in range(6)])

for i in range(len(prep)):
    prep_mark[int(prep[i][-1]) - 1][marks[i] - 3] += 1
    group_mark[int(group[i][-1]) - 1][marks[i] - 3] += 1

fig, axs = plt.subplots(nrows=2, ncols=1)

prep_name = []
prep.sort(key= lambda x: int(x[-1]))
for i in prep:
  if i not in prep_name:
    prep_name.append(i)


group_name = []
group.sort(key= lambda x: int(x[-1]))
for i in group:
  if i not in group_name:
    group_name.append(i)



axs[0].barh(prep_name, prep_mark[:, 0], label=str(3), color=colors[0])
axs[1].barh(group_name, group_mark[:, 0], label=str(3), color=colors[0])

for i in range(1, 8):
    axs[0].barh(prep_name, prep_mark[:, i], left=sum_arr(i)[0], label=str(i + 3),
               color=colors[i])
    axs[1].barh(group_name, group_mark[:, i], left=sum_arr(i)[1], label=str(i + 3),
               color=colors[i])

plt.legend(bbox_to_anchor=(1, 1))
figure = plt.gcf()
figure.set_size_inches(10, 5)
axs[0].set_title('marks per prep', loc='center', pad=10)
axs[1].set_title('marks per group', loc='center', pad=10)
plt.subplots_adjust(hspace=0.3)
plt.show()
```
